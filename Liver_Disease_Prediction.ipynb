```python
Data2['Status'] = Data2['Status'].replace('C', 0)
Data2['Status'] = Data2['Status'].replace('CL', 1)
Data2['Status'] = Data2['Status'].replace('D', 1)
```

```python
keys = ['Ascites','Hepatomegaly','Spiders','Edema']
for col in keys:
    Data2[col] = Data2[col].replace('N', 0)
    Data2[col] = Data2[col].replace('Y', 1)
    
Data2['Sex'] = Data2['Sex'].replace('M', 1)
Data2['Sex'] = Data2['Sex'].replace('F', 0)

Data2['Drug'] = Data2['Drug'].replace('D-penicillamine', 1)
Data2['Drug'] = Data2['Drug'].replace('Placebo', 0)

Data2['Edema'] = Data2['Edema'].replace('S', 0)
```

```python
from scipy import stats
import numpy as np
z=np.abs(stats.zscore(Data2[['Bilirubin','Alk_Phos','SGOT','Tryglicerides','Prothrombin','Copper','Albumin','Cholesterol']]))
print(z)
```

```python
## Removing Outliers
Data2_new=Data2[(z<3).all(axis=1)]
print('Before removing Outlier',Data2.shape)
print('After removing Outlier',Data2_new.shape)
```

```python
X1_train,X1_test=train_test_split(Data2_new,test_size=0.1)
```
